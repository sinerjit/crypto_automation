```python
from data_analysis import read_data, filter_data, calculate_trend_analysis, print_coin_data, calculate_price_volatility
from data_collector import run_data_collection  # Veri toplama fonksiyonunu çağır

# Veri dosyası yolları
plain_file_path = r'C:\Users\sarp_\Desktop\X\Data Analyst\Data_Analyst_Projects\crypto_automation\data\plain_crypto_data.csv'
clean_file_path = r'C:\Users\sarp_\Desktop\X\Data Analyst\Data_Analyst_Projects\crypto_automation\data\clean_crypto_data.csv'

# Ana fonksiyon
def main():
    while True:
        # Kullanıcıdan input al
        crypto_symbol = input("Please enter a crypto symbol (e.g., BTC) or 'All' for all data (type 'exit' to quit): ").strip()

        # Çıkış koşulu
        if crypto_symbol.lower() == 'exit':
            print("Exiting the program.")
            break

        # Veri toplama işlemi başlıyor
        print("Fetching and cleaning data...")
        run_data_collection(plain_file_path, clean_file_path)

        # Veri toplama işlemi tamamlandı
        print("Data collection completed. Now performing analysis...")

        # En güncel temizlenmiş veriyi oku
        df = read_data(clean_file_path)

        # Filtrelenmiş veriyi al
        filtered_data = filter_data(df, crypto_symbol)

        # Veri mevcut mu diye kontrol et
        if filtered_data.empty:
            print(f"No data found for symbol: {crypto_symbol.upper()}")
        else:
            # Ortalama yüzdesel değişimi hesapla
            average_percent_change = calculate_trend_analysis(filtered_data)
            
            # Fiyat volatilitesini hesapla
            price_volatility = calculate_price_volatility(filtered_data)

            # 'All' girildiyse tüm coin'leri sırayla yazdır
            if crypto_symbol.lower() == 'all':
                for i, row in filtered_data.iterrows():
                    print_coin_data(row, average_percent_change.iloc[i], price_volatility.iloc[i])
            else:
                # Tek bir coin için sonuçları yazdır
                row = filtered_data.iloc[0]
                # Sadece bir coin olduğu için average_percent_change ve price_volatility'nin ilk değeri alınır
                print_coin_data(row, average_percent_change.iloc[0], price_volatility.iloc[0])

# Eğer bu dosya doğrudan çalıştırılıyorsa ana fonksiyonu çağır
if __name__ == "__main__":
    main()
```
